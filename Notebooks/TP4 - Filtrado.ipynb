 $$ y(k) = \frac{1}{N} \sum_{i=0}^{N-1} x(k-i)  $$
 
 Para N = [3; 5]

   a) El diagrama de polos y ceros y su respuesta de módulo y fase


**Para N = 3 :**

$$ y(k) = \frac{1}{3} \sum_{i=0}^{2} x(k-i)  $$
 
$$ y(k) = \frac {1}{3} [x(k) + x(k-1) + x(k-2)] $$

$$ Y(z) = \frac {1}{3} [X(z) + {z}^{-1} X(z) + {z}^{-2} X(z)] $$

$$ Y(z) = \frac {1}{3} X(z) [1 + {z}^{-1} + {z}^{-2}] $$

$$ H(z) = \frac {Y(z)}{X(z)} = \frac {1}{3} \frac{1}{{z}^{2}} [{z}^{2} + z + 1] $$

$$ H(z) = \frac {Y(z)}{X(z)} = \frac {1}{3} \frac{{z}^{2} + z + 1}{{z}^{2}} $$

De donde se deduce que $ H(z) $ tiene dos polos al origen y dos ceros, uno en $(-\frac{1}{2} + j\frac{\sqrt(2)}{2})$ y el otro en $(-\frac{1}{2} - j\frac{\sqrt(2)}{2})$   

**Para N = 5 :**

$$ y(k) = \frac{1}{5} \sum_{i=0}^{4} x(k-i)  $$
 
$$ y(k) = \frac {1}{5} [x(k) + x(k-1) + x(k-2) + x(k-3) + x(k-4)] $$

$$ Y(z) = \frac {1}{5} [X(z) + {z}^{-1} X(z) + {z}^{-2} X(z) + {z}^{-3} X(z) + {z}^{-4} X(z)] $$

$$ Y(z) = \frac {1}{5} X(z) [1 + {z}^{-1} + {z}^{-2} + {z}^{-3} + {z}^{-4}] $$

$$ H(z) = \frac {Y(z)}{X(z)} = \frac {1}{5} \frac{1}{{z}^{5}} [{z}^{4} + {z}^{3} + {z}^{2} + z + 1] $$

$$ H(z) = \frac {Y(z)}{X(z)} = \frac {1}{5} \frac{{z}^{4} + {z}^{3} + {z}^{2} + z + 1}{{z}^{4}} $$

$ H(z) $ tiene cuatro polos al origen y cuatro ceros en (0.31 + j 0.95), (0.31 - j 0.95), (-0.81 + j 0.59) y (-0.81 + j 0.59)

c) Diseñe **al menos** dos filtros FIR y dos IIR para su comparación. Verifique que la respuesta en frecuencia responda a la plantilla de diseño
  

```python

##########################################
# Acá podés generar los gráficos pedidos #
##########################################


fs=1000

mat_struct = sio.loadmat('ECG_TP4.mat')
#
ecg_one_lead =mat_struct['ecg_lead'].flatten()
N = len(ecg_one_lead)

#Media movil de 10 muestras para reducir el ruido
ecgfil=np.convolve(ecg_one_lead,np.ones(10)/10)

data =ecgfil

#Periodograma de welch para analizar el espectro
#w = np.hamming(1024)
#f, pW = sig.welch(data, fs=fs, nfft=N, window='hanning', nperseg=int(np.round(N/2)) )


nyq_frec = fs / 2


# filter design
ripple = 0.5 # dB
atenuacion = 40 # dB

ws1 = 0.3 #Hz
wp1 = 0.5 #Hz
wp2 = 30.0 #Hz
ws2 = 35.0 #Hz




frecs = np.array([0.0,         ws1,         wp1,     wp2,     ws2,         nyq_frec   ]) / nyq_frec

gains = np.array([-atenuacion, -atenuacion, -ripple, -ripple, -atenuacion, -atenuacion])
gains = 10**(gains/20)

#Filtrols IIR
bp_sos_butter = sig.iirdesign(wp=np.array([wp1, wp2]) / nyq_frec, ws=np.array([ws1, ws2]) / nyq_frec, gpass=0.5, gstop=40., analog=False, ftype='butter', output='sos')
bp_sos_cauer = sig.iirdesign(wp=np.array([wp1, wp2]) / nyq_frec, ws=np.array([ws1, ws2]) / nyq_frec, gpass=0.5, gstop=40., analog=False, ftype='ellip', output='sos')

cant_coef = 501

#Filtros FIR
num_win =   sig.firwin2(cant_coef, frecs, gains , window='blackmanharris' )

den = 1.0

w, h_butter = sig.sosfreqz(bp_sos_butter)
_, h_cauer = sig.sosfreqz(bp_sos_cauer)

plt.figure()
plt.plot (w[0:int(len(w)/4)]/np.pi,np.abs(h_butter[0:int(len(w)/4)]),label='Butterworth')
plt.plot (w[0:int(len(w)/4)]/np.pi,np.abs(h_cauer[0:int(len(w)/4)]),label='Cauer')
axes_hdl = plt.gca()
axes_hdl.legend()


#Aplicación del filtro a la señal
ecgbutt = sig.sosfiltfilt(bp_sos_butter, data)
ecgcauer = sig.sosfiltfilt(bp_sos_cauer, data)
start=472000
stop=473000
plt.figure()
plt.plot(ecg_one_lead[start:stop],label='señal')
plt.plot(ecgfil[start:stop],label='media movil')
plt.plot(ecgbutt[start:stop],label='butter')
plt.plot(ecgcauer[start:stop],label='cauer')
axes_hdl = plt.gca()
axes_hdl.legend()
```

a) Estimación del movimiento de línea de base a partir del filtrado de mediana. A partir de la siguiente estimación del movimiento de línea de base 

$$ \hat{b} = \mathop{med}_{600} \{ \mathop{med}_{200} \{s\} \} $$

Donde $s$ representa la señal de ECG a partir de la que estimamos el movimiento de la línea de base $ \hat{b}$. El operador $\mathop{med}_{200}$ calcula muestra a muestra la mediana en los 200 ms previos a la muestra actual. Luego dicha estimación la substraemos para estimar la actividad eléctrica cardíaca $\hat{x}$

$$ \hat{x} = s - \hat{b} $$

<div class="alert alert-block alert-success">
<b>Ayuda:</b> Puede utilizar la función de interpolación <a href="https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.interpolate.CubicSpline.html">spline <b>scipy.interpolate.CubicSpline</b></a>
